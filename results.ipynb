```python
# Occupancy function -- Returns mean and stdev of time (s) in each zone
def occupancy(folder_path, date, exclude=None):
    occupancies = {'L': [], 'M': [], 'R': []}
    exclude = exclude or []
    ids = []
    for file_name in [f for f in os.listdir(folder_path) if f.endswith('.csv') and any(d in f for d in date) and not any(ex in f for ex in exclude)]:
        file_path = os.path.join(folder_path, file_name)
        print(file_name)
        ids.append(file_name[10:13])
        
        # Read and filter .csv file
        df = pd.read_csv(file_path, header=[2], index_col=0)
        df_snout = df.filter(like='snout').drop('snout.2', axis=1).drop('coords', axis=0).reset_index(drop=True).set_axis(['snout_x', 'snout_y'], axis=1)
       
        # Clean data
        df_snout['snout_x'] = pd.to_numeric(df_snout['snout_x'], errors='coerce')
        df_snout['snout_y'] = pd.to_numeric(df_snout['snout_y'], errors='coerce')
        df_snout.ffill(axis=1)
        df_snout.fillna(df_snout.median(), inplace=True)
        
        # Define ROIs
        frame_width = 320 
        frame_rate = 30
        left_line = frame_width // 3
        right_line = 2 * (frame_width // 3)
        time_in_L, time_in_M, time_in_R = 0, 0, 0
        
        # Iterate through each frame
        for i, row in df_snout.iterrows():
            x = row['snout_x']
            if x < left_line:
                time_in_L += 1
            elif x < right_line:
                time_in_M += 1
            else:
                time_in_R += 1
        
        # Append time spent in seconds
        time_in_L /= frame_rate
        time_in_M /= frame_rate
        time_in_R /= frame_rate
        occupancies['L'].append(time_in_L)
        occupancies['M'].append(time_in_M)
        occupancies['R'].append(time_in_R)
    '''
    # This checks for values/outliers by plotting
    for key, values in occupancies.items():
        plt.plot(range(1, len(values) + 1), values, label=key)
    plt.xlabel('Sample')
    plt.ylabel('Occupancy')
    plt.title('Value Check (occupancy)')
    plt.legend()
    '''
    means = [np.mean(occupancies['L']), np.mean(occupancies['M']), np.mean(occupancies['R'])]
    stds = [np.std(occupancies['L']), np.std(occupancies['M']), np.std(occupancies['R'])]
    return means, stds, occupancies['L'], occupancies['M'], occupancies['R'], ids
```

```python
# Velocity function -- Returns mean and stdev of velocity
def avg_velocity(folder_path, date, exclude=None):
    velocities = []
    exclude = exclude or []
    for file_name in [f for f in os.listdir(folder_path) if f.endswith('.csv') and any(d in f for d in date) and not any(ex in f for ex in exclude)]:
        file_path = os.path.join(folder_path, file_name)
        print(file_name)
        
        # Read and filter .csv file
        df = pd.read_csv(file_path, header=[2], index_col=0)
        df_snout = df.filter(like='snout').drop('snout.2', axis=1).drop('coords', axis=0).reset_index(drop=True).set_axis(['snout_x', 'snout_y'], axis=1)
       
        # Clean data
        df_snout['snout_x'] = pd.to_numeric(df_snout['snout_x'], errors='coerce')
        df_snout['snout_y'] = pd.to_numeric(df_snout['snout_y'], errors='coerce')
        df_snout.ffill(axis=1)
        df_snout.fillna(df_snout.median(), inplace=True)
        
        #Distance between frames
        df_snout['x_diff'] = df_snout['snout_x'].diff()
        df_snout['y_diff'] = df_snout['snout_y'].diff()
        df_snout['distance'] = np.sqrt(df_snout['x_diff']**2 + df_snout['y_diff']**2)

        # Calculate seconds between consecutive frames
        time_interval = 1 / 30  # 30 frames per second

        # Calculate velocity (distance moved divided by time interval)
        df_snout['velocity'] = df_snout['distance'] / time_interval
        # Average velocity
        velocities.append(df_snout['velocity'].mean())
    '''
    # This checks for values/outliers by plotting
    x_values = range(1, len(velocities) + 1)
    plt.plot(x_values, velocities)
    plt.xlabel('Sample')
    plt.ylabel('Velocity')
    plt.title('Value Check (velocity)')
    plt.legend()
    '''
    means = np.mean(velocities)
    stds = np.std(velocities)
    return means, stds, velocities
```

```python
# Function calculating number of bouts for specified behaviour

def calculate_bouts(frames):
    bouts = 0
    durations = []
    bout_start = frames[0]
    bout_end = frames[0]
    
    for frame in frames[1:]:
        if frame - bout_end <= 30:
            bout_end = frame
        else:
            bouts += 1
            durations.append(bout_end - bout_start + 1)
            bout_start = frame
            bout_end = frame
    
    bouts += 1
    durations.append(bout_end - bout_start + 1)

    return bouts, durations

def analyse_behaviour(frame_path, behaviour, date, exclude=None):
    all_durations = []
    all_bouts = []
    exclude = exclude or []
    ids=[]
    durtest=[]
    for file_name in [f for f in os.listdir(frame_path) if f.endswith('.npy') and any(d in f for d in date) and not any(ex in f for ex in exclude)]:
        file_path = os.path.join(frame_path, file_name)
        print(file_name)
        ids.append(file_name[10:13])
        
        array=np.load(file_path)
        beh_frames = retrieve_frames(array, behaviour)
        bouts, durations = calculate_bouts(beh_frames)
        all_durations.extend(durations)
        durtest.append(np.mean(durations))
        all_bouts.append(bouts)
    print(all_bouts)
    bout_means = np.mean(all_bouts)
    bout_stds = np.std(all_bouts)
    duration_means = np.mean(all_durations)
    duration_stds = np.std(durtest)
    
    return bout_means, bout_stds, duration_means, duration_stds, durtest, all_bouts, ids
```

```python
# Funtion to count transition probabilities

def transition_prob(behaviors):
    transition_counts = {} 
    
    # Count transitions
    for i in range(len(behaviors) - 1):
        current_behavior = behaviors[i]
        next_behavior = behaviors[i + 1]
        transition = (current_behavior, next_behavior)

        if transition in transition_counts:
            transition_counts[transition] += 1
        else:
            transition_counts[transition] = 1

    # Normalize counts to get probabilities
    transition_probabilities = {}
    total_transitions = sum(transition_counts.values())
    for transition, count in transition_counts.items():
        transition_probabilities[transition] = count / total_transitions

    return transition_probabilities
```

```python
# Transition Matrix function

behaviours2 = {0: 'Eat', 1: 'Eat', 2: 'Inv/Others', 3: 'Inv/Others', 4: 'Inv/Others', 5: 'Grooming', 6: 'Inv/Others', 7: 'Inv/Others', 8: 'Rest', 9: 'Locomotion', 10: 'Inv/Others'}
def analyse_transition(frame_path, date, exclude=None):
    all_prob = {}
    exclude = exclude or []
    for file_name in [f for f in os.listdir(frame_path) if f.endswith('.npy') and any(d in f for d in date) and not any(ex in f for ex in exclude)]:
        file_path = os.path.join(frame_path, file_name)
        print(file_name)
        
        array=np.load(file_path)
        annotated_array = [behaviours2.get(num, str(num)) for num in array]
        probabilities = transition_prob(annotated_array)
        
        # Aggregate transition probabilities from all files
        for transition, probability in probabilities.items():
            if transition in all_prob:
                all_prob[transition] += probability
            else:
                all_prob[transition] = probability
    
    normal_prob = {}
    total_prob = sum(all_prob.values())
    for transition, count in all_prob.items():
        normal_prob[transition] = count / total_prob
    print(normal_prob)
    states = list(set(transition[0] for transition in normal_prob.keys()))
    transition_matrix = np.zeros((len(states), len(states)))
    for i, state_i in enumerate(states):
        for j, state_j in enumerate(states):
            transition_probability = normal_prob.get((state_i, state_j), 0)
            transition_matrix[i, j] = transition_probability

    return transition_matrix
```

```python
# Retrieve box of interest

behaviours2 = {0: 'Eat', 1: 'Eat', 2: 'Inv/Others', 3: 'Inv/Others', 4: 'Inv/Others', 5: 'Grooming', 6: 'Inv/Others', 7: 'Inv/Others', 8: 'Rest', 9: 'Locomotion', 10: 'Inv/Others'}
def retrieve_transition(frame_path, date, from_state, to_state, exclude=None):
    values = []
    exclude = exclude or []
    for file_name in [f for f in os.listdir(frame_path) if f.endswith('.npy') and any(d in f for d in date) and not any(ex in f for ex in exclude)]:
        file_path = os.path.join(frame_path, file_name)
        print(file_name)
        
        array=np.load(file_path)
        annotated_array = [behaviours2.get(num, str(num)) for num in array]
        probabilities = transition_prob(annotated_array)
        
        key = (from_state, to_state)
        values.append(probabilities.get(key, 0))
    
    return values
```
